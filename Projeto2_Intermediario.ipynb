# Treinamento do Modelo

```python
# Import do TF e da ferramentas usadas
from __future__ import absolute_import, division, print_function, unicode_literals
import tensorflow as tf
from tensorflow.keras import layers

# Import de outras bibliotecas que serão usada
import numpy as np
import datetime
import os

# Imprimindo versão do TensorFlow
print(tf.__version__)
```

```python
X = df.drop('hazardous', axis=1)
X = X.drop('name', axis=1)
X = X.drop('orbiting_body', axis=1)
X = X.drop('sentry_object', axis=1)
y = df['hazardous']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0,test_size=0.2)
```

```python
modelo = tf.keras.Sequential()

modelo.add(layers.Flatten())
modelo.add(layers.Dense(800, kernel_initializer="random_uniform", bias_initializer="random_uniform", activation="sigmoid"))
modelo.add(layers.Dense(10, kernel_initializer="random_uniform", bias_initializer="random_uniform", activation="sigmoid"))

modelo.compile(optimizer="sgd", loss="sparse_categorical_crossentropy", metrics=["binary_accuracy"])
```

```python
log_dir = os.path.join( "logs", "fit", datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
results = modelo.fit(X_train, y_train, batch_size = 60000, epochs=99, callbacks=[tensorboard_callback])
```

Acurácia de 74,18%... tentaremos outra estratégia

```python
modelo = tf.keras.Sequential()

modelo.add(layers.Flatten())
modelo.add(layers.Dense(64, activation='relu'))
modelo.add(layers.Dense(10, activation='softmax'))

modelo.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
log_dir = os.path.join( "logs", "fit", datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
results = modelo.fit(X_train, y_train, epochs=20, callbacks=[tensorboard_callback])
```

84,42% é um resultado um pouco melhor... como podemos maximizá-lo?

```python
model = tf.keras.Sequential()
```

```python
#Input Layer
model.add(layers.Dense(units = 20, activation = 'relu'))
#layer1
model.add(layers.Dense(units = 10, activation = 'relu'))
#Layer2
model.add(layers.Dense(units = 5, activation = 'relu'))
#output Layer
model.add(layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
model.compile(optimizer= 'adam', loss = 'binary_crossentropy', metrics= ['accuracy'])
```

```python
results = model.fit(X_train, y_train,batch_size = 10, epochs= 5)
```

90,31% por mim é bem legal... vamos testar mais uma rede diferente, agora com dados mais filtrados

```python
model = tf.keras.Sequential()
model.add(layers.Dense(12, input_dim=6, activation='relu'))
model.add(layers.Dense(8, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()

# Fit Model

history = model.fit(X_train, y_train, batch_size = 18, epochs = 10,
    validation_split=0.1,verbose = 1,shuffle=True)
```

Última tentativa e muito satisfatória! 90,89%!
